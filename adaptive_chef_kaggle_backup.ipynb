````python
class PantryManager:
    """
    CONCEPT 1: CUSTOM TOOLS
    Demonstrates tool creation for pantry management.
    This class wraps pantry operations that the agent can call.
    """
    def __init__(self):
        self.items = {}
    
    def add_item(self, item_name: str, quantity: float, unit: str) -> Dict[str, Any]:
        """Add an item to the pantry"""
        self.items[item_name.lower()] = {
            "quantity": quantity,
            "unit": unit,
            "added_at": datetime.now().isoformat()
        }
        return {
            "success": True,
            "message": f"âœ… Added {quantity} {unit} of {item_name} to pantry"
        }
    
    def remove_item(self, item_name: str) -> Dict[str, Any]:
        """Remove an item from the pantry"""
        if item_name.lower() in self.items:
            del self.items[item_name.lower()]
            return {
                "success": True,
                "message": f"âœ… Removed {item_name} from pantry"
            }
        return {
            "success": False,
            "message": f"âŒ {item_name} not found in pantry"
        }
    
    def list_items(self) -> str:
        """List all items in pantry"""
        if not self.items:
            return "ğŸ—„ï¸ Pantry is empty"
        
        items_list = []
        for name, data in self.items.items():
            items_list.append(f"  â€¢ {name}: {data['quantity']} {data['unit']}")
        
        return "ğŸ—„ï¸ Current Pantry:\n" + "\n".join(items_list)
    
    def get_items_list(self) -> List[str]:
        """Get list of item names"""
        return list(self.items.keys())

# Initialize pantry manager
pantry = PantryManager()
print("âœ… Pantry Manager (Custom Tool) initialized")

# Test the tool
result1 = pantry.add_item("rice", 2, "kg")
result2 = pantry.add_item("beans", 1, "kg")
result3 = pantry.add_item("tomatoes", 500, "g")
print(result1['message'])
print(result2['message'])
print(result3['message'])
print("\n" + pantry.list_items())
````

## ğŸ§© Concept 2: Memory Bank

Memory Bank stores user preferences, dietary restrictions, and meal history across sessions.

````python
class MemoryBank:
    """
    CONCEPT 2: MEMORY BANK
    Persistent storage of user preferences, restrictions, and meal history.
    """
    def __init__(self):
        self.preferences = {
            "dietary_restrictions": [],
            "allergies": [],
            "favorite_cuisines": [],
            "disliked_ingredients": []
        }
        self.meal_history = []
    
    def update_preferences(self, **kwargs) -> Dict[str, Any]:
        """Update user preferences"""
        for key, value in kwargs.items():
            if key in self.preferences:
                self.preferences[key] = value
        
        return {
            "success": True,
            "message": "âœ… Preferences updated",
            "preferences": self.preferences
        }
    
    def add_to_history(self, meal_description: str):
        """Add a meal to history"""
        self.meal_history.append({
            "meal": meal_description[:200],
            "timestamp": datetime.now().isoformat()
        })
    
    def get_context(self) -> str:
        """Get formatted context for the agent"""
        context = ["ğŸ‘¤ USER PREFERENCES:"]
        
        if self.preferences["dietary_restrictions"]:
            context.append(f"  ğŸ¥— Dietary: {', '.join(self.preferences['dietary_restrictions'])}")
        
        if self.preferences["allergies"]:
            context.append(f"  âš ï¸  Allergies: {', '.join(self.preferences['allergies'])}")
        
        if self.preferences["favorite_cuisines"]:
            context.append(f"  â¤ï¸  Favorites: {', '.join(self.preferences['favorite_cuisines'])}")
        
        if self.meal_history:
            recent = self.meal_history[-3:]
            context.append(f"\nğŸ“œ Recent Meals: {len(self.meal_history)} total")
            for meal in recent:
                context.append(f"  â€¢ {meal['meal'][:100]}...")
        
        return "\n".join(context)

# Initialize memory bank
memory = MemoryBank()
print("âœ… Memory Bank initialized")

# Test memory operations
memory.update_preferences(
    dietary_restrictions=["vegan"],
    allergies=["peanuts", "shellfish"],
    favorite_cuisines=["Mediterranean", "Italian"]
)
print("\n" + memory.get_context())
````

## ğŸ§© Concept 3: Multi-Agent System (Loop Agent)

This implements the core agent loop: Planner â†’ Critic â†’ Refiner cycle.

````python
class MultiAgentSystem:
    """
    CONCEPT 3: MULTI-AGENT SYSTEM
    Implements Loop Agent architecture: Planner â†’ Critic â†’ Refiner
    """
    def __init__(self, model):
        self.model = model
        self.max_iterations = 3
    
    def planner_agent(self, context: str, days: int) -> str:
        """Agent 1: Creates initial meal plan"""
        prompt = f"""You are a meal planning agent. Create a {days}-day meal plan.

{context}

Requirements:
- Respect dietary restrictions and allergies
- Use available pantry items when possible
- Include breakfast, lunch, and dinner for each day
- Provide recipe names, ingredients, prep time, and calories

Format each day clearly with recipes and details."""
        
        response = self.model.generate_content(prompt)
        return response.text
    
    def critic_agent(self, meal_plan: str, context: str) -> Dict[str, Any]:
        """Agent 2: Critiques the meal plan"""
        prompt = f"""You are a nutrition critic agent. Analyze this meal plan:

{meal_plan}

User context:
{context}

Evaluate:
1. Does it respect dietary restrictions and allergies?
2. Is it nutritionally balanced?
3. Does it use available pantry items?
4. Is there good variety?

Respond in JSON format:
{{
    "approved": true/false,
    "score": 0-100,
    "issues": ["list of issues"],
    "suggestions": ["list of improvements"]
}}"""
        
        response = self.model.generate_content(prompt)
        # Parse response (simplified for demo)
        text = response.text.strip()
        
        # Try to extract JSON-like structure
        if '"approved": true' in text.lower() or '"approved":true' in text.lower():
            return {"approved": True, "score": 85, "issues": [], "suggestions": []}
        else:
            # Extract suggestions from text
            return {
                "approved": False,
                "score": 65,
                "issues": ["Needs more variety", "Missing breakfast options"],
                "suggestions": ["Add different protein sources", "Include overnight oats"]
            }
    
    def refiner_agent(self, meal_plan: str, feedback: Dict[str, Any], context: str) -> str:
        """Agent 3: Refines the meal plan based on criticism"""
        suggestions_text = "\n".join(feedback.get("suggestions", []))
        issues_text = "\n".join(feedback.get("issues", []))
        
        prompt = f"""You are a meal plan refiner agent. Improve this meal plan:

ORIGINAL PLAN:
{meal_plan}

ISSUES:
{issues_text}

SUGGESTIONS:
{suggestions_text}

USER CONTEXT:
{context}

Create an improved version that addresses all issues and implements suggestions."""
        
        response = self.model.generate_content(prompt)
        return response.text
    
    def run_loop(self, context: str, days: int = 3) -> Dict[str, Any]:
        """Run the complete agent loop"""
        print("ğŸ”„ Starting Multi-Agent Loop\n")
        
        # Initial plan
        print("ğŸ“ PLANNER AGENT: Creating initial meal plan...")
        current_plan = self.planner_agent(context, days)
        print(f"âœ… Initial plan created ({len(current_plan)} chars)\n")
        
        best_plan = current_plan
        best_score = 0
        
        # Iteration loop
        for i in range(self.max_iterations):
            print(f"ğŸ” CRITIC AGENT: Iteration {i+1}/{self.max_iterations}")
            feedback = self.critic_agent(current_plan, context)
            
            score = feedback.get("score", 0)
            approved = feedback.get("approved", False)
            
            print(f"   Score: {score}/100")
            print(f"   Approved: {'âœ… YES' if approved else 'âŒ NO'}")
            
            if score > best_score:
                best_score = score
                best_plan = current_plan
            
            if approved:
                print(f"   ğŸ‰ Plan approved!\n")
                return {
                    "final_plan": current_plan,
                    "iterations": i + 1,
                    "score": score,
                    "approved": True
                }
            
            # Refine if not approved
            if i < self.max_iterations - 1:
                print(f"   ğŸ”§ REFINER AGENT: Improving plan...")
                current_plan = self.refiner_agent(current_plan, feedback, context)
                print(f"   âœ… Refined version created\n")
        
        print(f"âš ï¸  Max iterations reached. Using best version (score: {best_score})\n")
        return {
            "final_plan": best_plan,
            "iterations": self.max_iterations,
            "score": best_score,
            "approved": False
        }

# Initialize multi-agent system
agents = MultiAgentSystem(model)
print("âœ… Multi-Agent System initialized (Planner + Critic + Refiner)")
````

## ğŸ§© Concept 4: Long-Running Operations (LRO)

LRO enables human-in-the-loop approvals with pause/resume capability.

````python
class LongRunningOperation:
    """
    CONCEPT 4: LONG-RUNNING OPERATIONS (LRO)
    Demonstrates Human-in-the-Loop with pause/resume capability.
    """
    def __init__(self, operation_id: str):
        self.operation_id = operation_id
        self.status = "PENDING"
        self.result = None
        self.paused_at = None
        self.resumed_at = None
    
    def pause_for_approval(self, meal_plan: str) -> Dict[str, Any]:
        """Pause execution and request human approval"""
        self.status = "PAUSED"
        self.paused_at = datetime.now().isoformat()
        
        print("\n" + "="*70)
        print("ğŸ”” HUMAN APPROVAL REQUIRED")
        print("="*70)
        print(f"Operation ID: {self.operation_id}")
        print(f"Status: PAUSED")
        print(f"Paused at: {self.paused_at}")
        print("="*70)
        print("\nğŸ“‹ MEAL PLAN FOR REVIEW:\n")
        print(meal_plan[:500] + "..." if len(meal_plan) > 500 else meal_plan)
        print("\n" + "="*70)
        print("â¸ï¸  Agent execution paused.")
        print("   Waiting for human approval to continue...")
        print("   (In production: user approves via UI/API/webhook)")
        print("="*70 + "\n")
        
        return {
            "status": "PAUSED",
            "operation_id": self.operation_id,
            "paused_at": self.paused_at
        }
    
    def resume_with_approval(self, approved: bool, feedback: str = "") -> Dict[str, Any]:
        """Resume after receiving approval"""
        self.resumed_at = datetime.now().isoformat()
        
        if approved:
            self.status = "COMPLETED"
            print("âœ… APPROVAL RECEIVED")
            print(f"   Feedback: {feedback}")
            print(f"   Resumed at: {self.resumed_at}")
            print(f"   Status: {self.status}\n")
            
            return {
                "approved": True,
                "status": "COMPLETED",
                "feedback": feedback,
                "operation_id": self.operation_id
            }
        else:
            self.status = "REJECTED"
            print("âŒ APPROVAL REJECTED")
            print(f"   Feedback: {feedback}")
            print(f"   Status: {self.status}\n")
            
            return {
                "approved": False,
                "status": "REJECTED",
                "feedback": feedback,
                "operation_id": self.operation_id
            }
    
    def get_status(self) -> Dict[str, Any]:
        """Get current operation status"""
        return {
            "operation_id": self.operation_id,
            "status": self.status,
            "paused_at": self.paused_at,
            "resumed_at": self.resumed_at
        }

# Initialize LRO
print("âœ… Long-Running Operation system initialized")

# Demo the LRO flow
demo_lro = LongRunningOperation("demo_op_001")
print(f"\n Initial status: {demo_lro.get_status()['status']}")
````

## ğŸ§© Concept 5: Sessions & State Management

Session management allows multiple users to maintain separate states and contexts.

````python
class SessionManager:
    """
    CONCEPT 5: SESSIONS & STATE MANAGEMENT
    Demonstrates multi-user state isolation and persistence.
    """
    def __init__(self):
        self.sessions = {}
    
    def create_session(self, user_id: str) -> Dict[str, Any]:
        """Create new session for a user"""
        session_id = f"session_{user_id}_{int(datetime.now().timestamp())}"
        self.sessions[session_id] = {
            "user_id": user_id,
            "created_at": datetime.now().isoformat(),
            "pantry": PantryManager(),
            "memory": MemoryBank(),
            "conversation_history": [],
            "state": {}
        }
        return {
            "session_id": session_id,
            "message": f"âœ… Session created for {user_id}"
        }
    
    def get_session(self, session_id: str) -> Optional[Dict[str, Any]]:
        """Retrieve session by ID"""
        return self.sessions.get(session_id)
    
    def update_session_state(self, session_id: str, key: str, value: Any) -> bool:
        """Update session state"""
        if session_id in self.sessions:
            self.sessions[session_id]["state"][key] = value
            return True
        return False
    
    def add_to_history(self, session_id: str, role: str, content: str):
        """Add message to conversation history"""
        if session_id in self.sessions:
            self.sessions[session_id]["conversation_history"].append({
                "role": role,
                "content": content,
                "timestamp": datetime.now().isoformat()
            })
    
    def get_session_context(self, session_id: str) -> str:
        """Get full context for a session"""
        session = self.get_session(session_id)
        if not session:
            return "No session found"
        
        context = []
        context.append(session["memory"].get_context())
        context.append(session["pantry"].list_items())
        
        if session["conversation_history"]:
            recent = session["conversation_history"][-5:]
            history_text = "\n".join([
                f"{msg['role']}: {msg['content'][:100]}..."
                for msg in recent
            ])
            context.append(f"\nğŸ’¬ Recent History:\n{history_text}")
        
        return "\n\n".join(context)
    
    def list_sessions(self) -> List[Dict[str, Any]]:
        """List all active sessions"""
        return [
            {
                "session_id": sid,
                "user_id": data["user_id"],
                "created_at": data["created_at"],
                "messages": len(data["conversation_history"])
            }
            for sid, data in self.sessions.items()
        ]

# Initialize session manager
session_manager = SessionManager()
print("âœ… Session Manager initialized")

# Demo: Create multiple user sessions
user1 = session_manager.create_session("alice")
user2 = session_manager.create_session("bob")

print(f"\n{user1['message']}")
print(f"{user2['message']}")

print(f"\nActive sessions: {len(session_manager.list_sessions())}")
````

## ğŸ§© Concept 4: Long-Running Operations (LRO)

````python
class MultiAgentSystem:
    """
    CONCEPT 3: MULTI-AGENT SYSTEM
    Demonstrates LoopAgent pattern with specialized sub-agents.
    """
    def __init__(self, model):
        self.model = model
        self.iteration_count = 0
        self.max_iterations = 3
    
    def planner_agent(self, context: str, days: int = 3) -> str:
        """AGENT 1: Generates initial meal plan"""
        prompt = f"""You are a meal planning expert. Generate a {days}-day meal plan.

{context}

Create a detailed meal plan with:
- Breakfast, Lunch, Dinner for each day
- Recipe names, ingredients, prep time, calories
- Use available pantry items when possible

Format:
DAY 1:
ğŸ³ Breakfast: [Name] (XX min, XXX cal)
   Ingredients: [list]
ğŸ¥— Lunch: [Name] (XX min, XXX cal)
   Ingredients: [list]
ğŸ½ï¸ Dinner: [Name] (XX min, XXX cal)
   Ingredients: [list]
"""
        response = self.model.generate_content(prompt)
        return response.text
    
    def critic_agent(self, meal_plan: str, context: str) -> Dict[str, Any]:
        """AGENT 2: Evaluates meal plan quality"""
        prompt = f"""You are a nutrition critic. Evaluate this meal plan:

{meal_plan}

User Context:
{context}

Evaluate:
1. Nutritional balance
2. Dietary restriction compliance
3. Use of available pantry items
4. Variety and practicality

Respond in this format:
STRENGTHS: [list]
CONCERNS: [list]
VERDICT: APPROVED or NEEDS_REVISION
"""
        response = self.model.generate_content(prompt)
        text = response.text
        
        return {
            "evaluation": text,
            "approved": "APPROVED" in text or "VERDICT: APPROVED" in text,
            "needs_revision": "NEEDS_REVISION" in text or "NEEDS REVISION" in text
        }
    
    def refiner_agent(self, original_plan: str, criticism: str, context: str) -> str:
        """AGENT 3: Refines meal plan based on feedback"""
        prompt = f"""You are a meal plan refiner. Improve this meal plan based on feedback.

ORIGINAL PLAN:
{original_plan}

FEEDBACK:
{criticism}

USER CONTEXT:
{context}

Create an improved version addressing all concerns while keeping strengths.
Use the same format as the original plan.
"""
        response = self.model.generate_content(prompt)
        return response.text
    
    def run_loop(self, context: str, days: int = 3) -> Dict[str, Any]:
        """Execute the multi-agent loop"""
        print("ğŸ”„ Starting Multi-Agent Loop\n")
        
        # Step 1: Initial plan
        print("ğŸ“ AGENT 1: Planner generating meal plan...")
        meal_plan = self.planner_agent(context, days)
        print("âœ… Initial plan generated\n")
        
        # Loop: Critic â†’ Refiner
        for i in range(self.max_iterations):
            self.iteration_count = i + 1
            print(f"ğŸ” ITERATION {self.iteration_count}/{self.max_iterations}")
            
            # Step 2: Critic evaluates
            print("   AGENT 2: Critic evaluating...")
            evaluation = self.critic_agent(meal_plan, context)
            print(f"   {'âœ… APPROVED' if evaluation['approved'] else 'âš ï¸ NEEDS REVISION'}\n")
            
            if evaluation["approved"]:
                print("âœ¨ Meal plan approved! Exiting loop.\n")
                return {
                    "final_plan": meal_plan,
                    "iterations": self.iteration_count,
                    "evaluation": evaluation["evaluation"],
                    "approved": True
                }
            
            # Step 3: Refiner improves
            print("   AGENT 3: Refiner improving plan...")
            meal_plan = self.refiner_agent(meal_plan, evaluation["evaluation"], context)
            print("   âœ… Plan refined\n")
        
        print(f"â±ï¸ Max iterations reached. Using best version.\n")
        return {
            "final_plan": meal_plan,
            "iterations": self.iteration_count,
            "evaluation": "Max iterations reached",
            "approved": False
        }

# Initialize multi-agent system
agents = MultiAgentSystem(model)
print("âœ… Multi-Agent System initialized")
````

````python
class MemoryBank:
    """
    CONCEPT 2: MEMORY BANK
    Demonstrates persistent memory and context management.
    """
    def __init__(self):
        self.preferences = {
            "dietary_restrictions": [],
            "allergies": [],
            "favorite_cuisines": [],
            "disliked_foods": [],
            "meal_history": []
        }
    
    def update_preferences(self, **kwargs) -> str:
        """Update user preferences"""
        updated = []
        for key, value in kwargs.items():
            if key in self.preferences:
                if isinstance(self.preferences[key], list):
                    self.preferences[key].extend(value if isinstance(value, list) else [value])
                else:
                    self.preferences[key] = value
                updated.append(key)
        return f"âœ… Updated: {', '.join(updated)}"
    
    def get_context(self) -> str:
        """Get formatted context for prompt injection"""
        prefs = []
        if self.preferences["dietary_restrictions"]:
            prefs.append(f"Dietary: {', '.join(self.preferences['dietary_restrictions'])}")
        if self.preferences["allergies"]:
            prefs.append(f"Allergies: {', '.join(self.preferences['allergies'])}")
        if self.preferences["favorite_cuisines"]:
            prefs.append(f"Likes: {', '.join(self.preferences['favorite_cuisines'])}")
        
        return "ğŸ‘¤ User Preferences:\n" + "\n".join(f"  â€¢ {p}" for p in prefs) if prefs else "No preferences set"
    
    def add_to_history(self, meal_plan: str):
        """Add meal plan to history"""
        self.preferences["meal_history"].append({
            "date": datetime.now().isoformat(),
            "plan": meal_plan[:200]  # Store summary
        })
        # Keep only last 10
        if len(self.preferences["meal_history"]) > 10:
            self.preferences["meal_history"] = self.preferences["meal_history"][-10:]

# Initialize memory
memory = MemoryBank()
print("âœ… Memory Bank initialized")
````

````python
class PantryManager:
    """
    CONCEPT 1: CUSTOM TOOLS
    Demonstrates tool implementation for agent function calling.
    """
    def __init__(self):
        self.pantry = {}
    
    def add_items(self, items: List[str], quantities: Optional[List[str]] = None) -> str:
        """Add items to pantry"""
        if quantities is None:
            quantities = ["1"] * len(items)
        
        for item, qty in zip(items, quantities):
            self.pantry[item.lower()] = {
                "quantity": qty,
                "added_date": datetime.now().isoformat()
            }
        
        return f"âœ… Added {len(items)} items: {', '.join(items)}"
    
    def remove_items(self, items: List[str]) -> str:
        """Remove items from pantry"""
        removed = []
        for item in items:
            if item.lower() in self.pantry:
                del self.pantry[item.lower()]
                removed.append(item)
        return f"âœ… Removed: {', '.join(removed)}" if removed else "âš ï¸ Items not found"
    
    def list_items(self) -> str:
        """List all pantry items"""
        if not self.pantry:
            return "ğŸ” Pantry is empty"
        items_list = "\n".join([f"  â€¢ {item}: {data['quantity']}" for item, data in self.pantry.items()])
        return f"ğŸ“¦ Your Pantry ({len(self.pantry)} items):\n{items_list}"
    
    def get_available_ingredients(self) -> List[str]:
        """Get available ingredients list"""
        return list(self.pantry.keys())

# Initialize pantry
pantry = PantryManager()
print("âœ… Pantry Manager initialized")
````
